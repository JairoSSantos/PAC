```python
def autocorr(x, m=True):
    result = np.correlate(x, x, mode='full')
    return result[result.size//2:] if m else result
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 4))
for ax in axs.flat: ax.axis('off')

for i in np.random.randint(0, len(img), 10):
    v = far_v[i]
    h = far_h.T[i]
    vt = far_v_th[i]
    ht = far_h_th.T[i]
    axs[0][0].plot(autocorr(v), 'k-', alpha=0.2)
    axs[0][1].plot(autocorr(h), 'k-', alpha=0.2)
    axs[1][0].plot(autocorr(vt), 'k-', alpha=0.2)
    axs[1][1].plot(autocorr(ht), 'k-', alpha=0.2)

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

freq = np.fft.fftfreq(len(img))
freq_filter = freq > 0
freq = freq[freq_filter]

Ux, Uy = [], []
Ax, Ay = [], []
Sx, Sy = 0, 0
N = 100
for i in np.random.randint(0, len(img), N): 
    u = norm(far_v[i])
    p = norm(far_h.T[i])
    
    ufft_complex = np.fft.fft(u)[freq_filter]
    pfft_complex = np.fft.fft(p)[freq_filter]
    
    ufft = np.abs(ufft_complex)
    pfft = np.abs(pfft_complex)
    
    Sx += ufft
    Sy += pfft
    
    u_peaks = np.isin(ufft, np.sort(ufft)[-2:])
    p_peaks = np.isin(pfft, np.sort(pfft)[-2:])
    
    Ax.append(list(np.angle(ufft_complex[u_peaks])))
    Ay.append(list(np.angle(pfft_complex[p_peaks])))

    Ux.append(list(freq[u_peaks]))
    Uy.append(list(freq[p_peaks]))
    
    axs[0][0].plot(freq, ufft, 'k-', alpha=2/N)
    axs[1][0].plot(freq, pfft, 'k-', alpha=2/N)
    
Ux = np.array(Ux)
axs[0][1].hist(Ux.flat)
ax = mode(np.round(Ux[:, 0], 5))[0][0]
bx = mode(np.round(Ux[:, 1], 5))[0][0]

print(ax, bx)

Uy = np.array(Uy)
axs[1][1].hist(Uy.flat)
ay = mode(np.round(Uy[:, 0], 5))[0][0]
by = mode(np.round(Uy[:, 1], 5))[0][0]

print(ay, by)
    
plt.show()
```

$$
\frac{u(x)}{dx} = -\alpha\tan(\omega x + \delta)
$$

$$
u(x) = \frac{\alpha}{\omega}\ln|\cos(\omega x + \delta)| + C
$$

```python
alpha = 1e-1
C = 1
f = lambda x, omega, delta: np.clip(alpha/omega*np.log(np.abs(np.cos(omega*x + delta))) + C, 0, 0.5*C)
```

```python
X, Y = np.meshgrid(np.arange(-128, 128), np.arange(-128, 128))
Gx = f(X, ax*np.pi, 1/ax)
Gy = f(Y, ay*np.pi, 1/ay)
color = [0, 1, 0]
plt.imshow(np.clip(
    gray2rgb(img) + 
    color*gray2rgb(rotate(norm(-Gy), -a1, mode='reflect'))+
    color*gray2rgb(rotate(norm(-Gx), -a1, mode='reflect')),
0, 1), cmap='gray')
```
