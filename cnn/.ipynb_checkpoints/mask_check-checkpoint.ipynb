```python
import os
from data import *
from preprocess import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.color import label2rgb
from skimage.measure import shannon_entropy
from scipy import stats
```

```python
is_jpg = lambda x: x.split('.')[-1] == 'jpg'
is_tif = lambda x: x.split('.')[-1] == 'tif'
folder = 'dataset_original'
files = os.listdir(folder)
imgs = np.stack([get(jpg_file, folder) for jpg_file in filter(is_jpg, files)])
masks = np.stack([get(tif_file, folder) for tif_file in filter(is_tif, files)]).astype(bool)
samples = [img[mask] for img, mask in zip(imgs, masks)]
back = [img[~mask] for img, mask in zip(imgs, masks)]
```

```python
i = np.random.randint(len(imgs))
plt.imshow(label2rgb(masks[i], imgs[i]))
```

```python
A = list(map(np.mean, samples))
A = list(map(np.mean, back))
bins = np.histogram(np.hstack((A, B)), bins=20)[1]
plt.hist(A, bins, alpha=0.6)
plt.hist(B, bins, alpha=0.6)
plt.show()
```

```python
A = list(map(np.mean, samples))
A = list(map(np.mean, back))
bins = np.histogram(np.hstack((A, B)), bins=20)[1]
plt.hist(A, bins, alpha=0.6)
plt.hist(B, bins, alpha=0.6)
plt.show()
```

```python
plt.hist([shannon_entropy(arr) for arr in samples], bins=20, alpha=0.6)
plt.hist([shannon_entropy(arr) for arr in back], bins=20, alpha=0.6)
plt.show()
```

```python
A = [stats.kurtosis(arr[~np.isnan(arr)]) for arr in samples]
B = [stats.kurtosis(arr[~np.isnan(arr)]) for arr in back]
bins = np.histogram(np.hstack((A, B)), bins=20)[1]
plt.hist(A, bins, alpha=0.6)
plt.hist(B, bins, alpha=0.6)
plt.show()
```
