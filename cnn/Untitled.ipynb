```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
quadro2 = pd.DataFrame({
    'Vm':[2, 5, 8.04, 10.05, 15.03], # tensão no voltímetro
    'escala':[1, 2, 5, 5, 10], # escala no osciloscópio
    'ndiv':[5+3/5, 7, 4+3/5, 5+3/5, 4+1/5], # número de divsões
})
quadro2['erro_Vm'] = quadro2.Vm*0.007 + 0.003
quadro2['Vpp'] = quadro2.escala*quadro2.ndiv
quadro2['Vp'] = quadro2.Vpp/2
quadro2['Vef'] = quadro2.Vp/np.sqrt(2)
quadro2['erro_Vef'] = quadro2.escala/10
quadro2['desvio_abs'] = np.abs(quadro2.Vm - quadro2.Vef)
quadro2['desvio_rel'] = quadro2.desvio_abs/quadro2.Vm*100
quadro2
```

```python
quadro3 = pd.DataFrame({
    'onda':['senoide']*2 + ['quadrada']*2 + ['triangular']*2,
    'f_gerador':[100, 5e3, 250, 1200, 600, 10e3],
    'escala':[1e-3, 20e-6, 0.2e-3, 0.1e-3, 0.1e-3, 5e-6],
    'ndiv':[2*(4+4/5), 10, 2*(9+4/5), 2*(4+1/5), 16, 4*(4+4/5)]
})
quadro3['periodo'] = quadro3.escala*quadro3.ndiv
quadro3['erro_periodo'] = quadro3.escala/10
quadro3['f'] = 1/quadro3.periodo
quadro3['erro_f'] = quadro3.erro_periodo/quadro3.periodo**2
quadro3['desvio_abs'] = np.abs(quadro3.f_gerador - quadro3.f)
quadro3['desvio_rel'] = quadro3.desvio_abs/quadro3.f_gerador*100
quadro3
```

```python
quadro6 = pd.DataFrame({
    'Vm':[4.66, 6.35, 1.73],
    'escala':[2, 5, 1],
    'ndiv':[6+3/5, 3+3/5, 4+4/5]
})
quadro6['erro_Vm'] = quadro2.Vm*0.007 + 0.003
quadro6['Vpp'] = quadro6.escala*quadro6.ndiv
quadro6['Vp'] = quadro6.Vpp/2
quadro6['Vef'] = quadro6.Vp/np.sqrt(2)
quadro6['delta'] = quadro6.escala/10
quadro6['erro_abs'] = np.abs(quadro6.Vm - quadro6.Vef)
quadro6['erro_rel'] = quadro6.erro_abs/quadro6.Vm*100
quadro6
```
