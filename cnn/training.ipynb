```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import os
```

# Importando os dados

```python
INFO_PATH = os.path.join('dataset', 'dataset.csv')
IMAGES_PATH = os.path.join('dataset', 'images')
```

```python
ds = pd.read_csv(INFO_PATH)
ds
```

```python
images = np.array([np.asarray(keras.utils.load_img(os.path.join(IMAGES_PATH, filename), color_mode='grayscale'))
               for filename in ds.filename])
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 8))

for ax, (_, area, im) in zip(axs.flat, ds.sample(len(axs.flat)).values):
    ax.imshow(im, cmap='gray')
    ax.set_title(area)
    
plt.show()
```

# Treinamento

```python
model = keras.Sequential((
    keras.layers.Rescaling(1/255, input_shape=(256, 256, 1)),
    keras.layers.Conv2D(16, 3, padding='same', activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Conv2D(32, 3, padding='same', activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Conv2D(64, 3, padding='same', activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(1, activation='linear')
))
model.compile(optimizer='adam', loss='mean_absolute_percentage_error')
```

```python
history = model.fit(images, ds.area, batch_size=40, validation_split=0.3, epochs=30)
```

```python
loss = history.history['loss']
plt.plot(range(len(loss)), loss)
plt.plot(range(len(loss)), history.history['val_loss'])
plt.show()
```
